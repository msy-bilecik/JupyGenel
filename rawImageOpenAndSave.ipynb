```python
"""
*.mhd 
"""



import SimpleITK as sitk
import matplotlib.image as img
from matplotlib import cm
from glob import glob
import os

basePath="D:\\datasets\\lung\\vessel(kaggle version)\\masks\\"
targetPath="D:\\datasets\\lung\\vessel(kaggle version)\\export\\masks\\"

tag="mask"
# print(basePath)
if not os.path.exists(targetPath):
    os.mkdir(targetPath)

all_MHDfile=sorted(glob(os.path.join(basePath,'*.mhd')))

for item in all_MHDfile:
    print(item)
    ct_scans = sitk.GetArrayFromImage(sitk.ReadImage(item, sitk.sitkInt16))
    base= os.path.splitext(os.path.basename(item))[0]
    saveFolder="{path}{base}{folder}".format(path=targetPath,folder="_IMG",base=base)
    if not os.path.exists(saveFolder):
        os.mkdir(saveFolder)
    
    for i in range(ct_scans.shape[0]):
        savePath="{saveFolder}\\{base}_{tag}_{num:04}.jpg".format(saveFolder=saveFolder,base=base,tag=tag,num=i)
        print(savePath)
        img.imsave(savePath,ct_scans[i],cmap = cm.gray)
```

```python
import matplotlib.image as img
from glob import glob
import os

img_source_path = "D:\\datasets\\ms\\miccai2008v\\flair\\export\\"
mask_source_path = "D:\\datasets\\ms\\miccai2008v\\lesion\\export\\"
img_target_path = "D:\\datasets\\ms\\miccai2008v\\fillSlice\\flair\\"
mask_target_path = "D:\\datasets\\ms\\miccai2008v\\fillSlice\\lesion\\"
maskPatch="GTmask"
imgPatch="slice"
print(img_source_path)
print(img_target_path)

if not os.path.exists(img_target_path):
   os.makedirs(img_target_path)
if not os.path.exists(mask_target_path):
   os.makedirs(mask_target_path)

img_files_list = []
for root, directories, files in os.walk(img_source_path):
   for name in files:
      img_files_list.append(os.path.join(root, name))
# img_files_list=[x.lower() for x in img_files_list]

mask_files_list = []
for root, directories, files in os.walk(mask_source_path):
   for name in files:
      mask_files_list.append(os.path.join(root, name))
# print(mask_files_list)
# mask_files_list = [x.lower() for x in mask_files_list]

print(len(img_files_list), len(mask_files_list))

i=0

for mask_file_name in mask_files_list:
   mask_img = img.imread(mask_file_name)
   mask_weight=sum(sum(sum(mask_img)))
   if mask_weight>50:
      img_file_name = mask_file_name.replace("lesion", "flair", 1).replace(
            "lesion", "FLAIR", 2).replace(maskPatch, imgPatch)

      if img_file_name in img_files_list:
            i=i+1
            slice_img = img.imread(img_file_name)
            base_img_file_name = os.path.splitext(
                os.path.basename(img_file_name))[0].replace("_FLAIR_","_")+".jpg"
            base_mask_file_name = os.path.splitext(
                os.path.basename(img_file_name))[0].replace("_FLAIR_slice_","_GTmask_")+".jpg"
            target_imgFile_Path = img_target_path+"\\"+base_img_file_name
            target_maskFile_Path = mask_target_path+"\\"+base_mask_file_name
            img.imsave(target_imgFile_Path,slice_img)
            img.imsave(target_maskFile_Path, mask_img)

print(i)

```

```python
import matplotlib.image as img
import os
import glob
import cv2

root_dir = 'D:\\datasets\\ms\\isbi\\isbi200x\\valMask\\'
mask_root_dir = 'D:\\datasets\\ms\\isbi\\flairPNG2\\'
mask_target_dir = 'D:\\datasets\\ms\\isbi\\isbi200x\\val\\'

all_images = sorted(glob.glob(os.path.join(root_dir, '*.jpg')))
# display(all_images)
for item in all_images:
    print("\n", item)
    fname = os.path.basename(item)
    maskFname = fname.replace("GTmask","slice")
    readMask = mask_root_dir+maskFname
    print("readMask:", readMask)
    mask = cv2.imread(readMask)
#     display(mask)
    writeMask = mask_target_dir+maskFname
    print("wMask:", writeMask)
    img.imsave(writeMask, mask)
```
